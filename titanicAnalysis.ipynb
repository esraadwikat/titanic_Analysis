```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load Excel File (By_Class sheet)
df = pd.read_excel("Titanic_Analysis_Cleaned_with_Charts.xlsx", sheet_name="By_Class")
```

```python
# Step 2: Check the data structure
print("\u001b[1mData Preview:\u001b[0m")
print(df.head())
print("\nColumns:", df.columns)
```

```python
# Step 3: Ensure survival percentage is on the right scale (convert to 0–100 if needed)
df['Survived%'] = df['Survived%'] * 100  # Already in 0–1 format
```

```python
# Step 4: Add death percentage column (if not present)
df['Died%'] = df['Died'] / df['Total'] * 100
```

```python
# Step 5: Sort by Pclass (to keep order 1 → 3)
df = df.sort_values('Pclass')
```

```python
# Step 6: Plot survival rate by class
plt.figure(figsize=(6, 4))
plt.bar(df['Pclass'], df['Survived%'], color='skyblue')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Percentage')
plt.xticks([1, 2, 3])
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Step 7: Compare survival and death percentages
plt.figure(figsize=(6, 4))
plt.plot(df['Pclass'], df['Survived%'], marker='o', label='Survived%', color='green')
plt.plot(df['Pclass'], df['Died%'], marker='o', label='Died%', color='red')
plt.title('Survival vs Death Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Percentage')
plt.xticks([1, 2, 3])
plt.ylim(0, 100)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Step 8: Summary/Interpretation for the teacher
analysis_text = """
Interpretation:

This analysis shows a clear relationship between passenger class and survival rate.
- First-class passengers had the highest survival rate (~47%).
- Second and third-class passengers had lower survival rates (~32% and ~33%).
- The death rate increased in lower classes.

This supports the idea that passengers in higher classes had better access to lifeboats and assistance, possibly due to proximity or priority treatment.
"""

print(analysis_text)
```
